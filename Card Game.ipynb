# Card Game Analysis

A card game is played between two players. In total there are 25 rounds in one game.The rules of the games are as follows
<br>1-A player wins a game if his sum of cards is greater than his opponent's sum of cards</br>
<br>2-The game will stop if one of the player runs of the cards</br>
<br>3-If in a round , a player runs out of the cards the cards are given back to the player and player with the greater sum wins and the game stops</br>
<br>4-The game ends after 25 rounds</br>



A deck is created that consists of 36 cards and sum of the deck is also calculated.

```python
##To create a Deck , so a game can be played among two opponents.
deck = list(range(2,11))*4
print(deck)
print(f'Length of deck:{len(deck)}')
```

```python
## To calculate the total number of points available in the deck.
total=sum(deck)
print(f'Total points of the deck:{total}')
```

Imported random Library is used to shuffle the deck so that the cards can be distributed. The function shuffle is used to shuffle the elements. After that cards are distributed among the players

```python
## Distribute equal number of cards to both the players
import random
def distribution (deck):
    p1=[]
    p2=[]
    random.shuffle(deck)
    for i in range(len(deck)):
        if i%2==0:
            p1.append(deck[i])
        else:
            p2.append(deck[i])
    #print(len(player_1))
    #print(len(player_2))
    return(p1,p2)
```

Below statement are used to print the no of cards and the sum that the player has.

```python
## Invoke the function inorder to distribute the cards equally among both the players. 

cards_distributed=distribution(deck)
player_1_original=cards_distributed[0]
player_2_original=cards_distributed[1]
print(f'Overall cards that are distributed to both players:{cards_distributed}')
print(f'The cards distributed to player 1:{player_1_original}')
print(f'Player 1 has {len(player_1_original)} cards')
print(f'The cards distributed to player 2:{player_2_original}' )
print(f'Player 2 has {len(player_2_original)} cards')

total_player_1=sum(player_1_original)
total_player_2=sum(player_2_original)
print(f'Total Sum of cards of player 1 : {total_player_1}')
print(f'Total Sum of cards of player 2: {total_player_2}')
```

```python
def analysis(deck):
    game_val=1
    winning_set=[]
    losing_set=[]
    lowest_val_set=[]
    count_set=[]
    while(game_val<=50):
        cards_distributed=distribution(deck)
        player_1=cards_distributed[0]
        player_2=cards_distributed[1]
        game_new_returns=game_new(player_1,player_2)
        winning_value=game_new_returns[0]
        lowest_val=game_new_returns[1]
        count=game_new_returns[2]
        winning_set.append(winning_value)
        losing_set.append(sum(deck)-winning_value)
        lowest_val_set.append(lowest_val)
        count_set.append(count)
        game_val+=1
    return((winning_set,losing_set,lowest_val_set,count_set))
    
```
